#Merge files

Once the data has been cleaned it will be merged to allow for more complex analysis techniques on the dataset as a whole.

Pension and Social Welfare dataframes above contain 3 identical columns as shown below. This means the pension column values can be merged to the social welfare dataset.

```python
pension['area'].equals(social_welfare['area'])
```

```python
pension['county'].equals(social_welfare['county'])
```

```python
pension['area_code'].equals(social_welfare['area_code'])
```

```python
pension_vals = pension['pension'].values.tolist()
```

```python
social_welfare['pension'] = pension_vals
social_welfare.head(3)
```

This process is repeated for long distance income, gross income and short distance income.

```python
long_distance_income_vals = long_distance_income['Median Income'].values.tolist()
```

```python
social_welfare['Long_Dist_Income'] = long_distance_income_vals
social_welfare.head(3)
```

```python
data = social_welfare
```

```python
income = gross_income['Income'].values.tolist()
data['gross_income'] = income
data.head(3)
```

```python
sd_income = short_distance_income['income'].values.tolist()
data['short_dist_income'] = sd_income
data.head(3)
```

Income section is then merged to this dataframe.

```python
income_section.head(3)
```

```python
merged_df = data.merge(income_section, how = 'inner', on = ['area_code', 'area', 'county'])
merged_df.head(3)
```

Social Welfare column is oved so that area information comes first

```python
sw_col = merged_df.pop('Social Welfare')
merged_df.insert(7, 'Social Welfare', sw_col)
merged_df.head(3)
```

#Final Data

Once the above dataset has been created all that is left to merge is the census data.

```python
census_data = saps_data
census_data.head(3)
```

As the dataframe above shows this dataset's dimension is 3409x802. This is a very large dataset that will cause long runtimes on any future algorithgms. This dataset has its dimensions reduced by combining similar columns (ages) into range values rather than individual values.

```python
t1_data = census_data.loc[:, census_data.columns.str.contains('T1')]
```

```python
female_age_data = t1_data.iloc[:, 35: 70]
```

```python
female_age_data.columns
```

Ages 0-19 are grouped.

```python
female_age_data['F0-5'] = female_age_data.iloc[:,0:5].sum(axis=1)
female_age_data['F6-10'] = female_age_data.iloc[:,6:10].sum(axis=1)
female_age_data['F11-15'] = female_age_data.iloc[:,11:15].sum(axis=1)
female_age_data['F16-20'] = female_age_data.iloc[:,16:20].sum(axis=1)
female_age_data = female_age_data.iloc[:, 20:]
female_age_data.head(3)
```

Data is sorted into ascending age order.

```python
female_age_data = female_age_data.reindex(sorted(female_age_data.columns), axis=1)
f6 = female_age_data.pop('F6-10')
female_age_data.insert(1, 'F6-10', f6)
female_age_data.head(3)
```

The above process is repeated on the male data

```python
male_age_data = t1_data.loc[:, :'T1_1AGETM']
male_age_data.head(3)
```

```python
male_age_data['M0-5'] = male_age_data.iloc[:,0:5].sum(axis=1)
male_age_data['M6-10'] = male_age_data.iloc[:,6:10].sum(axis=1)
male_age_data['M11-15'] = male_age_data.iloc[:,11:15].sum(axis=1)
male_age_data['M16-20'] = male_age_data.iloc[:,16:20].sum(axis=1)
male_age_data = male_age_data.iloc[:, 20:]
male_age_data.head(3)
```

```python
male_age_data = male_age_data.reindex(sorted(male_age_data.columns), axis=1)
male_age_data.head(3)
```

```python
male_age_data = male_age_data.reindex(sorted(male_age_data.columns), axis=1)
f6 = male_age_data.pop('M6-10')
male_age_data.insert(1, 'M6-10', f6)
male_age_data.head(3)
```

```python
census_data.head(3)
```

```python
male_female_age_data = pd.concat([male_age_data, female_age_data], axis=1)
male_female_age_data.head(3)
```

```python
a_data = census_data.iloc[:,3:73]
a_data.head(3)
```

The old census age data is removed and replaced with the grouped age data.

```python
census_data.drop(census_data.iloc[:,3:73], inplace=True, axis=1)
census_data.head(3)
```

```python
total_data = pd.concat([census_data, male_female_age_data], axis=1)
total_data.head(3)
```

```python
census_data = total_data
census_data.head(5)
```

The initial merged dataframe is merged with the census dataframe containing a single dataframe ciontaining all relevant data for this analysis.

```python
total_data = pd.concat([merged_df, census_data], axis=1)
total_data.head(3)
```
